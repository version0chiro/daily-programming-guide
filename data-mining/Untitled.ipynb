```python
import openpyxl
wb = openpyxl.load_workbook('../data/FINAL450.xlsx')
ws = wb.get_sheet_by_name('Sheet1')
# print(ws.cell(row=2, column=1).hyperlink.target)
```

```python
ws.cell(row=3,column=2).value
```

```python
weblinks = []
topics = []
statements = []
```

```python
for i in range(3,478):
    try:
        print(ws.cell(row=i,column=2).hyperlink.target)
        print(ws.cell(row=i,column=2).value)
        print(ws.cell(row=i,column=1).value)
        weblinks.append(ws.cell(row=i,column=2).hyperlink.target)
        topics.append(ws.cell(row=i,column=1).value)
        statements.append(ws.cell(row=i,column=2).value)
        
    except Exception as e:
        print(e)
        print("skipping row number ",i);
```

```python
weblinks
```

```python
webSeries = pd.Series(weblinks)
topicSeries = pd.Series(topics)    
statementSeries = pd.Series(statements)
```

```python
# create a dictonary
data = {"topic": topicSeries,
        "statement": statementSeries,
        "link": webSeries}
  
# Concatenating the series side
# by side as depicted by axis=1
# If you want to concatenate the 
# series one below the other
# change the axis to zero.
df = pd.concat(data,
               axis = 1)
```
